4)What is difference between ordered data type and unordered data type?

Ordered Data Type-Ordered data types maintain a specific order of elements.
                  The position or index of each element is significant and fixed.
                  Elements are accessed based on their position in the data structure.
                  It allows for accessing elements by their position, enabling operations like indexing,
                          slicing, and iterating in a predictable order.
                  Ex include lists, arrays, tuples, and strings.
                                   
Unordered Data Type-
                    Unordered data types do not have a fixed order of elements.
                    The position or index of elements is not meaningful.
                    Elements are accessed based on their values or keys.
                    It focus on fast lookup and retrieval of elements based on their values or keys, 
                          typically using hashing or other mechanisms to optimize search operations.
                    Ex include sets, dictionaries, and hash tables
